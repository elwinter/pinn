```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt("eplasma3.dat")
```

```python
# Extract and plot the boundary conditions at (t, x=0).
nt = 101
nx = 101
t = data[:nx, 0]
x = data[:nx, 1]
n1 = data[:nx, 2]
u1x = data[:nx, 3]
E1x = data[:nx, 4]
plt.plot(t, n1, label="$n_1$")
plt.plot(t, u1x, label="$u_{1x}$")
plt.plot(t, E1x, label="$E_{1x}$")
plt.legend()
plt.xlabel("t at x=0")
plt.ylabel("$n_1$, $u_{1x}$, $E_{1x}$")
plt.grid()
plt.show()
```

```python
# Extract and plot the initial conditions conditions at (t=0, x).
nt = 101
nx = 101
t = data[nx:, 0]
x = data[nx:, 1]
n1 = data[nx:, 2]
u1x = data[nx:, 3]
E1x = data[nx:, 4]
plt.plot(x, n1, label="$n_1$")
plt.plot(x, u1x, label="$u_{1x}$")
plt.plot(x, E1x, label="$E_{1x}$")
plt.legend()
plt.xlabel("x at t=0")
plt.ylabel("$n_1$, $u_{1x}$, $E_{1x}$")
plt.grid()
plt.show()
```
